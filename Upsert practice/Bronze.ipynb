```python
#importing packages
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS datamodeling.bronze
```

```python
if spark.catalog.tableExists("datamodeling.bronze.bronze_tabel"):
    last_load_date = spark.sql("SELECT max(order_date) FROM datamodeling.bronze.bronze_tabel").collect()[0][0]
else:
    last_load_date = '1000-01-01'
```

```python
last_load_date
```

```python

df = spark.sql(f"""SELECT * FROM datamodeling.default.source_data
WHERE order_date > '{last_load_date}'""")
```

```python
df.createOrReplaceTempView("Bronze_Source")
```

```python
%sql
select * from Bronze_Source;
```

```python
%sql
create or replace table datamodeling.bronze.bronze_tabel as select * from Bronze_Source;
```

```python
%sql
select * from datamodeling.bronze.Bronze_tabel
```
