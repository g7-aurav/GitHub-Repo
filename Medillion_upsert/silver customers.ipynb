```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS medillian_upsert.silver
```

```python
df = spark.read.table("medillian_upsert.bronze.customers")
df.display()
```

```python
df = df.drop("_rescued_data")
df.display()
```

```python
df = df.withColumn("domains",split(col("email"),"@")[1])
df.display()
```

```python
df = df.withColumn("full_name",concat(col("first_name"),lit(" "),col("last_name")))
df = df.drop("first_name","last_name")
df.display()
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("medillian_upsert.silver.customers")
```
